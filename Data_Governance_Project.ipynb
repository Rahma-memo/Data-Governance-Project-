```python
import pandas as pd
import pandera as pa
from pandera import Column, DataFrameSchema, Check

# Assuming df is your DataFrame
df = pd.read_csv('/content/framingham.csv', encoding='utf-8')

schema = pa.DataFrameSchema(
    {
        "male": Column(int, Check.isin([0, 1])),
        "age": Column(int, Check.in_range(0, 120)),
        "education": Column(int, Check.isin(range(5))),  # Assuming education is encoded as integers from 0 to 4
        "currentSmoker": Column(int, Check.isin([0, 1])),
        "cigsPerDay": Column(int, Check.in_range(0, 60)),  # Assuming a person can't smoke more than 60 cigarettes a day
        "BPMeds": Column(int, Check.isin([0, 1])),
        "prevalentStroke": Column(int, Check.isin([0, 1])),
        "prevalentHyp": Column(int, Check.isin([0, 1])),
        "diabetes": Column(int, Check.isin([0, 1])),
        "totChol": Column(float, Check.in_range(100, 400)),  # Assuming total cholesterol level is between 100 and 400
        "sysBP": Column(float, Check.in_range(50, 250)),  # Assuming systolic blood pressure is between 50 and 250
        "diaBP": Column(float, Check.in_range(30, 150)),  # Assuming diastolic blood pressure is between 30 and 150
        "BMI": Column(float, Check.in_range(10, 50)),  # Assuming BMI is between 10 and 50
        "heartRate": Column(int, Check.in_range(40, 200)),  # Assuming heart rate is between 40 and 200 beats per minute
        "glucose": Column(float, Check.in_range(50, 300)),  # Assuming glucose level is between 50 and 300
        "TenYearCHD": Column(int, Check.isin([0, 1])),
    },
    coerce=True
)

validated_df = schema.validate(df)

print(validated_df)
```

**Generalization:** involves replacing specific values with more general or less precise values. It involves transforming or categorizing data into broader categories or ranges. For example, instead of storing exact ages, you might generalize ages into age groups (e.g., 20-30, 30-40, etc.).

```python
# Convert the 'age' column to a numerical data type
df['age'] = pd.to_numeric(df['age'], errors='coerce')

# Anonymize the 'age' column by grouping the values into ranges
bins = [30, 40, 50, 60, 70]
labels = [ '30-39', '40-49', '50-59', '60-70']
df['age'] = pd.cut(df['age'], bins=bins, labels=labels)
df
```

```python
# Convert the 'glucose' column to a numerical data type
df['glucose'] = pd.to_numeric(df['glucose'], errors='coerce')

# Anonymize the 'glucose' column by grouping the values into ranges
bins = [40, 100, 200, 300, 400]
labels = ['40-99', '100-199', '200-299', '300-400']
df['glucose'] = pd.cut(df['glucose'], bins=bins, labels=labels)
df
```

**Perturbation:**  involves adding random noise to sensitive data to make it harder to recognize. For example, a person's salary might be increased or decreased by a small amount to protect their privacy. The amount of noise added can be adjusted to balance privacy with data utility.

```python
dfp = pd.read_csv('/content/framingham.csv')

def add_noise(dfp, column, std = None):   #std=standard deviation
    if std == None:
        std = dfp[column].std()

    withNoise = dfp[column].add(np.random.normal(0, std, dfp.shape[0]))
    return withNoise
# List of columns to be anonymized
columns = ["glucose"]

# Apply noise addition on the specified columns
for column in columns:
    dfp[column] = add_noise(dfp, column)

dfp["glucose"]
```

# **Data Pseudonymization**

**Hashing:** involves converting data into a fixed-size string of bytes using a hash function

```python
import pandas as pd
import hashlib

df_hashing = pd.read_csv('/content/framingham.csv')

# Define a function to hash data
def hash_data(data):
    return hashlib.sha256(str(data).encode()).hexdigest()

# Iterate through each column (excluding non-numeric columns)
for column in df_hashing.columns:
    if pd.api.types.is_numeric_dtype(df_hashing[column]):
        # Hash the column using a lambda function
        df_hashing['Hashed_' + column] = df_hashing[column].apply(lambda x: hash_data(x))

hashed_columns = [col for col in df_hashing.columns if col.startswith('Hashed_')]
hashed_df = df_hashing[hashed_columns]
hashed_df
```

**Encryption:** involves transforming data into a ciphertext using an encryption algorithm and a secret key.

```python
import pandas as pd
from cryptography.fernet import Fernet

df_hash= pd.read_csv('/content/framingham.csv')
# Generate a key
key = Fernet.generate_key()

# Initialize cipher
cipher = Fernet(key)

# Define a function to encrypt data
def encrypt_data(data):
    return cipher.encrypt(str(data).encode()).decode()

# Iterate through each column
for column in df_hash.columns:
    # Encrypt the column using a lambda function
    df_hash['Encrypted_' + column] = df_hash[column].apply(encrypt_data)


encrypted_columns = [col for col in df_hashing.columns if col.startswith('Encrypted_')]
encrypted_df = df_hashing[hashed_columns]
encrypted_df
```

**Attribute-Based Access Control (ABAC)**

```python
import pandas as pd

# Create a sample dataframe
data = {
    "user_id": [1, 2, 3],
    "name": ["Max", "Nina", "Alice"],
    "user roles": ["admin", "data owner", "viewer"]
}

users_df = pd.DataFrame(data)
print(users_df)
```

```python
from py_abac import PDP, Policy, AccessRequest
from py_abac.storage.memory import MemoryStorage

def check_access(user_df, user_name, action, user_role):
    """
    Check if access is allowed based on user attributes.
    Args:
        user_df (pd.DataFrame): Users dataframe.
        user_name (str): User's name.
        action (str): Action (e.g., "create", "delete", "update").
        user_role (str): User's role.

    Returns:
        bool: True if access is allowed, False otherwise.
    """
    allowed_users = ["Max", "Nina"]
    allowed_actions = ["create", "delete", "update"]

    if user_name in allowed_users and action in allowed_actions and user_role == "data owner":
        return True
    else:
        return False

# Example access request
user_name = "Max"
action = "update"
user_role = "data owner"

# Check if access is allowed
if check_access(users_df, user_name, action, user_role):
    print(f"Access granted for user '{user_name}' to perform action '{action}'.")
else:
    print(f"Access denied for user '{user_name}' to perform action '{action}'.")

```

```python
# Create an in-memory policy
policy_json = {
    "uid": "1",
    "description": "Max and Nina are allowed to create, delete, update any resources only if the user role matches.",
    "effect": "allow",
    "rules": {
        "subject": [
            {"$.name": {"condition": "Equals", "value": "Max"}},
            {"$.name": {"condition": "Equals", "value": "Nina"}}
        ],
        "resource": {"$.name": {"condition": "RegexMatch", "value": ".*"}},
        "action": [
            {"$.method": {"condition": "Equals", "value": "create"}},
            {"$.method": {"condition": "Equals", "value": "delete"}},
            {"$.method": {"condition": "Equals", "value": "update"}}
        ],
        "context": {"$.role": {"condition": "CIDR", "value": "data owner"}}
    },
    "targets": {},
    "priority": 0
}

# Create a policy object
policy = Policy.from_json(policy_json)

# Set up in-memory policy storage
storage = MemoryStorage()
storage.add(policy)

# Create a Policy Decision Point (PDP)
pdp = PDP(storage)

# Example access request
request_json = {
    "subject": {"id": "", "attributes": {"name": "Habiba"}},
    "resource": {"id": "", "attributes": {"name": "myrn:example.com:resource:123"}},
    "action": {"id": "", "attributes": {"method": "update"}},
    "context": {"role": "data owner"}
}

# Create an access request object
request = AccessRequest.from_json(request_json)

# Check if access request is allowed
if pdp.is_allowed(request):
    print(f"Access granted for user '{user_name}' to perform action '{action}'.")
else:
    print(f"Access denied for user '{user_name}' to perform action '{action}'.")
```
